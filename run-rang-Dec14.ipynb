```python
def avg_PD(seed):
    pd = []
    for i in seed:
        residue_n = i[2:]
        AA = pdb_df[pdb_df['residue_number']==int(residue_n)]['residue_name'].tolist()[0]
        pd.append(property_distance(AA, i[0]))
    return sum(pd)/len(pd)

def mimo_avg(input_list):
    if input_list == []:
        return None
    avg_list = []
    for seed in input_list:
        pd = []
        for i in seed:
            residue_n = i[2:]
            AA = pdb_df[pdb_df['residue_number']==int(residue_n)]['residue_name'].tolist()[0]
            pd.append(property_distance(AA, i[0]))
        avg_list.append(sum(pd)/len(pd))
    return sum(avg_list)/len(avg_list)

def order(A):
    if A == sorted(A,reverse=False):
        return True
    elif A == sorted(A,reverse=True):
        return True
    else:
        return False
```
